```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio files
audio_file_1, sr_1 = librosa.load(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Data\Audio data\Noise reduction\sample with noise.wav")
audio_file_2, sr_2 = librosa.load(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Data\Audio data\Noise reduction\sample without noise.wav")

# Extract the Mel spectrogram for both audio files
spec_1 = librosa.feature.melspectrogram(y=audio_file_1, sr=sr_1)
spec_2 = librosa.feature.melspectrogram(y=audio_file_2, sr=sr_2)

# Plot the Mel spectrogram for both audio files
plt.subplot(2, 1, 1)
librosa.display.specshow(spec_1, y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram - Audio file with noise')

plt.subplot(2, 1, 2)
librosa.display.specshow(spec_2, y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram - Audio file without noise')

plt.tight_layout()
plt.show()
```
