```python
from mpl_toolkits import mplot3d
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\N\t20052025002020.csv")
df2 = pd.read_csv(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\NS\t20052025002023.csv")
df3 = pd.read_csv(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\RS\t20052025002043.csv")
df4 = pd.read_csv(r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\UT\t20052025002029.csv")

# Create a list of your dataframes
df_list = [df1, df2, df3, df4]

# Create a list of your labels
labels = ['Normal screw', 'No-screw', 'Robot slip', 'Under-tightened screw']

# Initialize the plot
fig = plt.figure(figsize=(12,12))
ax = plt.axes(projection='3d')

# Set labels for axes
ax.set_xlabel('$TCP_x$', fontsize=10, rotation=150)
ax.set_ylabel('$TCP_y$')
ax.set_zlabel('$TCP_z$', fontsize=10, rotation=120)

for i, df in enumerate(df_list):
    zline = df["TCP_x (mm)"]
    xline = df["TCP_y (mm)"]
    yline = df["TCP_z (mm)"]

    ax.plot3D(xline, yline, zline, linewidth=2, label=labels[i])

ax.view_init(20, 40)
ax.legend()

plt.savefig("AllPlots.png", dpi=300)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# list of files

files = [r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\N\t20052025002020.csv",
         r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\NS\t20052025002023.csv",
         r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\RS\t20052025002043.csv",
         r"C:\Users\AMRAN\OneDrive - Aalborg Universitet\8. semester\Projekt\CODE\Machine Learning\Final data\Task data\UT\t20052025002029.csv"]

plt.figure(figsize=(10,5))

# choose a color cycle for the plots, one color for each file
colors = ['b', 'g', 'r', 'c', 'm']

for file, color in zip(files, colors):
    df = pd.read_csv(file)
    # plot the torque as a function of time
    foldername = os.path.basename(os.path.dirname(file))  # get folder name
    plt.plot(df['Time (ms)']/1000, df['Robot_I (A)'], color=color, label=foldername)

plt.xlabel('Time (s)', fontsize=14)  # increased font size
plt.ylabel('Robot_I (A)', fontsize=14)  # increased font size
plt.title('Robot Current over time for different classes', fontsize=16)  # increased font size

# increase x-tick and y-tick label sizes
plt.tick_params(axis='both', which='major', labelsize=12)

# add a box around the plot
#plt.box(on=True)

# add a grid
plt.grid(True)

# add a legend
plt.legend()
plt.show()
```
