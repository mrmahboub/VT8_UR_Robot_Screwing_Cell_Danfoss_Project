```python
#Add label to the files inside of a specific folder


import os
import pandas as pd

specific_folder_path = r"C:\Users\GHB\Desktop\Final test\Task data\N"

# Get the folder name (label)
label = os.path.basename(specific_folder_path)

# Iterate through the CSV files in the specific folder
for file_name in os.listdir(specific_folder_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(specific_folder_path, file_name)
        
        # Read the CSV file using pandas
        df = pd.read_csv(file_path)
        
        # Add a new column called "label" and fill it with the folder name
        df["label"] = label
        
        # Save the modified CSV file, overwriting the original file
        df.to_csv(file_path, index=False)
```

```python
#Add label to the files inside all of the subfolders in a specified folder


import os
import pandas as pd

intrinsic_folder_path = r"C:\Users\GHB\Desktop\Test dataset\Intrinsic data"

# Iterate through the subfolders in the input folder
for folder_name in os.listdir(intrinsic_folder_path):
    subfolder_path = os.path.join(intrinsic_folder_path, folder_name)
    
    # Check if the current path is a directory
    if os.path.isdir(subfolder_path):
        label = folder_name
        
        # Iterate through the CSV files in the subfolder
        for file_name in os.listdir(subfolder_path):
            if file_name.endswith(".csv"):
                file_path = os.path.join(subfolder_path, file_name)
                
                # Read the CSV file using pandas
                df = pd.read_csv(file_path)
                
                # Add a new column called "label" and fill it with the subfolder name
                df["label"] = label
                
                # Save the modified CSV file, overwriting the original file
                df.to_csv(file_path, index=False)
```
